```python
import numpy as np
A = np.array([[1, 2],
[3, 4]])
B = np.array([[5, 6],
[7, 8]])
v = np.array([2, 4])
R=A*B
print("element-wise multiplication A * B \n",R);
R=A@B
print("matrix multiplication A @ B \n",R)
R=A*v
print("A * v using broadcasting \n",R)
```
