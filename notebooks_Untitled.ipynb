```python
import os
os.makedirs("lab4_data/input", exist_ok=True)
os.makedirs("lab4_data/output", exist_ok=True)
os.makedirs("lab4_data/logs", exist_ok=True)
for item in os.listdir("lab4_data"):
    path = os.path.join("lab4_data", item)
    
    if os.path.isdir(path):
        print(item, "-> Folder")
    else:
        print(item, "-> File")

os.rename("lab4_data/logs", "lab4_data/history")
```

```python
file_path = "lab4_data/input/notes.txt"

with open(file_path, "w") as f:
    f.write("Line 1\n")
    f.write("Line 2\n")
    f.write("Line 3\n")
    f.write("Line 4\n")
    f.write("Line 5\n")
```

```python
with open(file_path, "r") as f:
    lines = f.readlines()

print("Number of lines:", len(lines))
```

```python
source = "lab4_data/input/notes.txt"
destination = "lab4_data/output/summary.txt"

with open(source, "r") as src:
    content = src.read()

with open(destination, "w") as dest:
    dest.write(content)
```
