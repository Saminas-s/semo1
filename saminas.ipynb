```python
age = int(input("Enter your age: "))
if age < 13:
    print("Child")
elif age <= 19:
    print("Teen")
elif age <= 59:
    print("Adult")
else:
    print("Senior")
```

```python
num = int(input("Enter a number: "))
if num > 0:
    print("Positive")
else:
    print("Not Positive")
if 10 <= num <= 50:
    print("Number is between 10 and 50")
else:
    print("Number is NOT between 10 and 50")
```
