```python
print("----- Question 1 -----")
try:
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    result = num1 / num2
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")
except ValueError:
    print("Error: Invalid input. Please enter numbers only.")
else:
    print("Result =", result)
print("\n----- Question 2 -----")
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def update_age(self, new_age):
        self.age = new_age
    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)
p = Person("Saminas", 19)
p.update_age(20)
p.display()
print("\n----- Question 3 -----")
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
class Manager(Employee):
    def add_bonus(self, bonus):
        self.salary += bonus
    def display(self):
        print("Name:", self.name)
        print("Total Salary:", self.salary)
m = Manager("Saminas", 5000)
m.add_bonus(1000)
m.display()
print("\n----- Question 4 -----")
class Student:
    def __init__(self):
        self.name = input("Enter name: ")
        try:
            age_input = input("Enter age: ")
            age = int(age_input)
            if age < 0:
                raise ValueError
            self.age = age
        except ValueError:
            print("Error: Age must be a positive number.")
            self.age = None
    def display(self):
        if self.age is not None:
            print("Name:", self.name)
            print("Age:", self.age)
        else:
            print("Invalid data entered.")
s = Student()
s.display()
```
