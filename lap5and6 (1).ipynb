```python

numbers = []
for i in range(5):
    num = float(input(f"Enter number {i+1}: "))
    numbers.append(num)
total = sum(numbers)
average = total / len(numbers)
maximum = max(numbers)
minimum = min(numbers)
print(f"\nSum: {total}")
print(f"Average: {average}")
print(f"Maximum: {maximum}")
print(f"Minimum: {minimum}")
```

```python
words = []
for i in range(5):
    word = input(f"Enter word {i+1}: ")
    words.append(word)
words_tuple = tuple(words)
print(f"\nFirst word: {words_tuple[0]}")
print(f"Last word: {words_tuple[-1]}")
print(f"Total number of words: {len(words_tuple)}")
search_word = input("\nEnter a word to search for: ")
if search_word in words_tuple:
    print(f"The word '{search_word}' exists in the tuple")
else:
    print(f"The word '{search_word}' does not exist in the tuple")
```

```python
students = {}
for i in range(3):
    name = input(f"Enter student {i+1} name: ")
    score = float(input(f"Enter {name}'s score: "))
    students[name] = score
print("\nAll students and their scores:")
for name, score in students.items():
    print(f"{name}: {score}")
highest_student = max(students, key=students.get)
highest_score = students[highest_student]
print(f"\nStudent with highest score: {highest_student} ({highest_score})")
```

```python
list1 = input("Enter numbers for the first list (space-separated): ").split()
list2 = input("Enter numbers for the second list (space-separated): ").split()
set1 = set(map(int, list1))
set2 = set(map(int, list2))
union = set1 | set2
intersection = set1 & set2
difference = set1 - set2
print(f"\nFirst set: {set1}")
print(f"Second set: {set2}")
print(f"Union: {union}")
print(f"Intersection: {intersection}")
print(f"Difference (first - second): {difference}")
```

```python
names = []
for i in range(5):
    name = input(f"Enter name {i+1}: ")
    names.append(name)

# Convert list to tuple
names_tuple = tuple(names)
print(f"\nNames tuple: {names_tuple}")

# Create dictionary with name lengths
name_lengths = {name: len(name) for name in names}
print("Name lengths dictionary:")
for name, length in name_lengths.items():
    print(f"{name}: {length} characters")

# Create set of unique name lengths
unique_lengths = set(name_lengths.values())
print(f"\nUnique name lengths set: {unique_lengths}")
```
